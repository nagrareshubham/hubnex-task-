```python
import pandas as pd
import numpy as np
```

```python
d1 = pd.read_excel('C:/Users/User/Downloads/ECOM.xlsx', sheet_name = ['customer', 'Product', 'Sales'])
d1
```

```python
top50_customer = d1['customer'].head(50).to_dict(orient='records')

top50_sales = d1['Sales'].head(50).to_dict(orient='records')

top50_product = d1['Product'].head(50).to_dict(orient='records')
top50_customer 
top50_sales
top50_product
```

```python
customer_data = pd.DataFrame(top50_customer)
print()
customer_data
sales_data = pd.DataFrame(top50_sales)
print()
sales_data
product_data = pd.DataFrame(top50_product)
product_data
```

```python
customer_data.to_csv('customer_top50.csv', index=0)
sales_data.to_csv('sales_top50.csv', index=0)
product_data.to_csv('product_top50.csv', index=0)
```

```python
with pd.ExcelWriter('ecom_top50_data.xlsx') as writer:
    customer_data.to_excel(writer, sheet_name='customer_top50', index=0)
    sales_data.to_excel(writer, sheet_name='Sales_top50', index=0)
    product_data.to_excel(writer, sheet_name='product_top50', index=0)
```
