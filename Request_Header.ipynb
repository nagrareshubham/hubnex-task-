```python
cUrl = 'https://worldtimeapi.org/api/timezone/Asia/Kolkata'
cParams = {
    "datetime" : "today",
    "day_of_week" : "Wednesday",
    "day_of_year" : 150,
    "timezone" : "IST"
}
cHeader = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
```

```python
cResp = rq.get(cUrl, params=cParams, headers= cHeader)
```

```python
print(cResp.json())
```
